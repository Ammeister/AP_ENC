```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
Image.MAX_IMAGE_PIXELS = 400000000
```

```python
img = Image.open("./monomodal.png")
img0 = Image.open("./monomodal_c0.png")
img1 = Image.open("./monomodal_c1.png")
img2 = Image.open("./monomodal_c2.png")
img3 = Image.open("./monomodal_c3.png")
img4 = Image.open("./monomodal_c4.png")

fig, axs = plt.subplots(6, figsize = (10, 20))

axs[0].imshow(img)
axs[0].set_title("Graphe monoparti : Réseau des personnes par cooccurrence\ndans une organisation")

axs[1].imshow(img0)
axs[1].set_title("Graphe monoparti : intégration de la classe 0")

axs[2].imshow(img1)
axs[2].set_title("Graphe monoparti : intégration de la classe 1")

axs[3].imshow(img2)
axs[3].set_title("Graphe monoparti : intégration de la classe 2")

axs[4].imshow(img3)
axs[4].set_title("Graphe monoparti : intégration de la classe 3")

axs[5].imshow(img4)
axs[5].set_title("Graphe monoparti : intégration de la classe 4")

plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[]);

fig.tight_layout()

fig.savefig("./monomodal_compil.png", format="png", dpi=300)
```

```python
img0 = Image.open("../bimodal/bimodal_d5.png")
img1 = Image.open("./monomodal_d5.png")

fig, axs = plt.subplots(2, figsize = (10, 8))

axs[0].imshow(img0)
axs[0].set_title("Graphe biparti : personnes et organisations (degré ≥ 5)")

axs[1].imshow(img1)
axs[1].set_title("Graphe monoparti : Répartition des classes d'échevins")

plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[]);

fig.tight_layout()

fig.savefig("./biparti_monoparti_d5.png", format="png", dpi=300)
```
